```python
import matplotlib.pyplot as plt
from mlxtend.data import three_blobs_data
from mlxtend.cluster import Kmeans
import pandas as pd
import numpy as pd

X, y = three_blobs_data()
plt.scatter(X[:, 0], X[:, 1], c='white')
plt.show()
```

```python
df = pd.DataFrame(X)
df['y'] = y

print(df)

# writer = pd.ExcelWriter('C:/Users/Armando/Downloads/VarMxltendCluster.xlsx', engine='xlsxwriter')
# df.to_excel(writer, sheet_name='Info', index=False)
# writer.save()
```

```python


km = Kmeans(k=3,
            max_iter=50,
            random_seed=1,
            print_progress=3)

km.fit(X)

print('Iterations until convergence:', km.iterations_)
print('Final centroids:\n', km.centroids_)
```

```python
y_clust = km.predict(X)

plt.scatter(X[y_clust == 0, 0],
            X[y_clust == 0, 1],
            s=50,
            c='lightgreen',
            marker='s',
            label='cluster 1')

plt.scatter(X[y_clust == 1,0],
            X[y_clust == 1,1],
            s=50,
            c='orange',
            marker='o',
            label='cluster 2')

plt.scatter(X[y_clust == 2,0],
            X[y_clust == 2,1],
            s=50,
            c='lightblue',
            marker='v',
            label='cluster 3')


plt.scatter(km.centroids_[:,0],
            km.centroids_[:,1],
            s=250,
            marker='*',
            c='red',
            label='centroids')

plt.legend(loc='lower left',
           scatterpoints=1)
plt.grid()
plt.show()
```
