# Estimación del Modelo

```python
# Variables
y = df['TotalCost']
#X = df[['Output', 'FuelPrice', 'LoadFactor']]

X = df[['intercept','Output', 'FuelPrice', 'LoadFactor']]
```

```python
# Model
model = PanelOLS(dependent= y,
             exog=X,
             entity_effects=True, # Para indicar efectos fijos
             time_effects=False)  # para indicar efectos aleatorios

print(model.fit(cov_type='clustered', cluster_entity=True))
```
